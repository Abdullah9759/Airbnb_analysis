```python
import pandas as pd
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("compressed_data.csv")
```

```python
df.head()
```

```python
df.columns
```

# Checking missing value

```python
print(df.isnull().sum())
```

# handaling missing values

```python
df['last review']= pd.to_datetime(df['last review'], errors='coerce')
```

```python
df.info()
```

```python
df.fillna({'review per month':0,'last review':df['last review'].min()}, inplace=True)
```

```python
df.dropna(subset=['NAME','host name'],inplace=True)
```

```python
print(df.isnull().sum())
```

```python
df=df.drop(columns=['license','house_rules'],errors="ignore")
```

```python
df['price']=df['price'].replace('[\$,]','',regex=True).astype(float)
df['service fee']=df['service fee'].replace('[\$,]','',regex=True).astype(float)
```

```python
df.head()
```

# remove duplicates

```python
df.drop_duplicates(inplace=True)
```

# descriptive statistics

```python
df.describe()
```

# Visualization

```python
plt.figure(figsize=(20,6))
sns.histplot(df['price'] ,bins=50 ,kde=True, color="red")
plt.title('Distribution of listing price')
plt.xlabel('Price $')
plt.ylabel('Frequency')
plt.show()
```

```python
df['room type']
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x='room type', data=df, color='green')
plt.title('Room distribution')
plt.xlabel('room type')
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.countplot(y='neighbourhood group',data=df,color='yellow',order=df['neighbourhood group'].value_counts().index)
plt.title('Number of listing by neighbourhood')
plt.xlabel('Count')
plt.ylabel('Neighbourhood')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='room type', y='price', data=df ,hue='room type',palette="Set1")
plt.title('Price vs Room type')
plt.xlabel('Room type')
plt.ylabel('Price $')
plt.show()
```

```python
review_over_time=df.groupby(df['last review'].dt.to_period('M')).size()
```

```python
plt.figure(figsize=(20,6))
review_over_time.plot(kind='line',color='orange')
plt.title('Number of review over time')
plt.xlabel('Date')
plt.ylabel('Number of review')
plt.show()
```
